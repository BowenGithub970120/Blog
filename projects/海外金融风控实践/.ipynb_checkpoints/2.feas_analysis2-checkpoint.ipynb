### 特征选择

```python
# #   计算指标 
# 行缺失率 
from feas_select import  calc_feas,get_drop
def calc_feas(df,catefeas=[]):
    print(df.isnull().sum(axis=1).sort_values(ascending=False) / df.shape[1])

    # 特征缺失率
    miss_features = df.isnull().sum().sort_values(ascending=False) / df.shape[0]
    display(miss_features.head(20))

    # 特征单值率
    sigle_rate = {}
    for var in df.columns:
        sigle_rate[var]=(df[var].value_counts().max()/df.shape[0])
    print('single',sorted( sigle_rate.items(),key=lambda x:x[1], reverse=True)[0:20])

    # 方差
    var_features = df.var().sort_values()
    print('varance',var_features)

    #相关系数
    corr_df1=df.loc[:,set(df.columns )- set(catefeas)].corr() 
    upper1 = corr_df1.where(np.triu(np.ones(corr_df1.shape), k=1).astype(np.bool))
    catefeas = sorted(catefeas,reverse=True)  # 类别型sorted相关性
    from sklearn import preprocessing
    ##类别型先整数编码
    cdf=pd.DataFrame()
    for x in set(catefeas)&set(df.columns ):
        label = preprocessing.LabelEncoder()
        cdf[x] = label.fit_transform(df[x])
    corr_df2 = cdf.corr(method="kendall") 
    upper2 = corr_df2.where(np.triu(np.ones(corr_df2.shape), k=1).astype(np.bool))
    return upper1,upper2,var_features,miss_features,sigle_rate

upper1,upper2,var_features,miss_features,sigle_rate = calc_feas(df,catefeas)
```

```python
import matplotlib.pyplot as plt 
%matplotlib inline

def get_drop(upper1,upper2,var_features,miss_features,sigle_rate):
    # 筛选特征    
    drop_list = []


    miss_drop = miss_features[miss_features >  0.98].index.values #缺失率
    print('miss_drop',len(miss_drop))
    drop_list.extend(miss_drop)
    plt.plot(miss_features.values)
    plt.show()

    var_drop = var_features[var_features <= 0.001].index.values  #方差
    drop_list.extend(var_drop)
    plt.plot(var_features[var_features<10].values)
    plt.show()
    print('var_drop',len(var_drop))

    single_drop = [key for key, value in sigle_rate.items() if value > 0.98]#单值率
    drop_list.extend(single_drop)
    print('single_drop',len(single_drop))


    corr_drop1 = [column for column in upper1.columns if any(upper1[column].abs()>0.95)] #相关系数
    print('corr_drop1',len(corr_drop1))
    drop_list.extend(corr_drop1)
    aa = []
    for col in upper1.fillna(np.inf).values:
        aa.extend(col)
    plt.plot(sorted(aa))
    plt.show()

    corr_drop2 = [column for column in upper2.columns if any(upper2[column].abs()>0.95)] #相关系数
    print('corr_drop2',len(corr_drop2))
    drop_list.extend(corr_drop2)

    # drop_list.extend([])  # 无关特征

    return sorted(list(set(drop_list)))

drop_list = get_drop(upper1,upper2,var_features,miss_features,sigle_rate)
print(len(drop_list),drop_list)
df2 = df.drop(drop_list,axis=1)
df2.head()
```

```python
# 转为类别型、数值型
def to_category_or_num(df, category_list, drop_list):
    for ft in set(df.columns)-set(drop_list):
        if ft in category_list:
            df[ft] = df[ft].astype('category')
        else:
            df[ft] = pd.to_numeric(df[ft], errors='coerce')
            
to_category_or_num(df2, catefeas, ['ID','report_timestamp'])
```

### 对重要特征暴力特征加工ft 并筛选


```python

def featuretools_topfeas(basedf_file='./data/filter_feas_df.pkl', importance_file='./data/feature_importance.pkl', catefeas=[],join_key='ID',final_file='./data/top30_tofeaturetools.pkl',topn=50):
    feature_importance = pd.read_pickle(importance_file)
    df2 = pd.read_pickle(basedf_file)

    top_list = list(feature_importance.sort_values('split',ascending=False)[0:topn].sort_values('feature',ascending=True).feature.values)
    print(top_list)
    top_list.append(join_key)
    top20_df = df2[top_list]
    top20_df.head()

    # featuretools 特征 
    # ft.list_primitives()
    import gc
    import featuretools as ft

    gc.enable()
    gc.collect()
    es = ft.EntitySet(id='dfs')
    es.entity_from_dataframe(entity_id='top20_df',         # 增加一个数据框
                 dataframe=top20_df,
                 index=join_key)


    ## 运行特征衍生
    features_matrix2,feature_names = ft.dfs(entityset=es,
                                             target_entity='top20_df',
                                             trans_primitives=["divide_numeric",'multiply_numeric','add_numeric','subtract_numeric'],
                                             max_depth=3,n_jobs=-1,verbose=10)

    print(features_matrix2.shape)
    top_list.remove(join_key)
    features_matrix2 = features_matrix2.drop(top_list, axis=1)


    train_df3 = df2.merge(features_matrix2, how='left',left_on=join_key,right_index=True)
    del es,features_matrix2


    # 特征选择
    upper1,upper2,var_features,miss_features,sigle_rate = calc_feas(train_df3,catefeas)
    drop_list = get_drop(upper1,upper2,var_features,miss_features,sigle_rate)
    print(len(drop_list),'drop_list')
    train_df3 = train_df3.drop(drop_list,axis=1)

    train_df3.to_pickle(final_file)
    display(train_df3.head())
    gc.enable()
    gc.collect()
```
